```python
#Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
#Load and read dataset 
df = pd.read_csv('Melbourne_housing_FULL.txt')
df.head()
```

```python
df.nunique()
```

```python
df.shape
```

```python
useful_cols = ['Suburb','Propertycount', 'Rooms', 'Type', 'Method',
               'SellerG', 'Regionname','CouncilArea', 'Bedroom2','Distance','Bathroom','Car','Landsize' ,'BuildingArea', 'Price']

df = df[useful_cols]
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
```

```python
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
X = df.drop('Price', axis = 1)
```

```python
X
```

```python
y = df['Price']
```

```python
y.head()
```

```python
len(X)
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state = 42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.score(X_train, y_train)
```

#### L1 Regularization 

```python
from sklearn import linear_model

lasso_reg = linear_model.Lasso(alpha=50, max_iter= 100, tol =.1)
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_test, y_test)
```

```python
lasso_reg.score(X_train, y_train)
```

#### L2 Regularization 

```python
from sklearn.linear_model import Ridge
```

```python
ridge_reg = Ridge(alpha=50, max_iter= 100, tol =.1)
ridge_reg.fit(X_train, y_train)
```

```python
ridge_reg.score(X_test, y_test)
```

```python
ridge_reg.score(X_train, y_train)
```
